# Este programa, tiene como finalidad la creación de carpetas, selección y separación de archivos, para su orden y posterior trabajo con ellos.

```python
from selenium import webdriver
import time
import re
import os
import shutil
```

### Esta celda nos permite realizar la creación de una serie de directorios, en el path seleccionado

```python
os.makedirs('D:/documentos/Codigos aprendiaje/codigo abierto/ecobici/csvs/')
path = 'D:/documentos/Codigos aprendiaje/codigo abierto/ecobici/csvs/'
Anios = ['2019','2018','2017','2016','2015','2014','2013','2012','2011','2010','2009']

for i in Anios:
    try:
        os.makedirs(path+i)
        print("Directorio " , path+i ,  " Creado ")
    except Exception:
        print("Directorio " , path+i ,  " ya existia ")
```

### Enlistamos todos los archivos que se encuentren en el directorio

```python
files = []
for (path, dirnames, filenames) in os.walk('D:/Descargas'):
    files.extend(os.path.join(path, name) for name in filenames)

print (files)
```

### Filtramos los archivos que contengan la palabra csv

```python
#def endswith(list):
res = [k for k in files if 'csv' in k]
res    
```

### Si tenemos archivos duplicados, con expresiones regulares, los filtramos y los removemos de nuestros archivos.

```python
regex = re.compile('\).csv')
remove_files = list(filter(regex.search, res))
print(remove_files)
for r in range(len(remove_files)):
    if os.path.exists(remove_files[r]):
        os.remove(remove_files[r])
        res.remove(remove_files[r])
    else:
        print("El archivo no existe")
```

### Se mueven (cortar y pegar) todos los archivos a cada uno de sus directorios correpsondiente, filtrandolos por año.

```python
path = 'D:/documentos/Codigos aprendiaje/codigo abierto/ecobici/csvs/'
Anios = ['2019','2018','2017','2016','2015','2014','2013','2012','2011','2010','2009']
for ani in Anios:
    regex = re.compile(ani)
    selected_files = list(filter(regex.search, res))
    print(selected_files)
    for copy in range(len(selected_files)):
        try:
            shutil.move(selected_files[copy], path+ani)
            print(path+ani)
        except Exception as e:
            print(e)
```
